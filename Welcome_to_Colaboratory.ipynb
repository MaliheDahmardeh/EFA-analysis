#EFA for selected columns

```
selected_columns = ['E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'E7', 'E8', 'E9', 'E10',
                    'N1', 'N2', 'N3', 'N4', 'N5', 'N6', 'N7', 'N8', 'N9', 'N10',
                    'A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10',
                    'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10',
                    'O1', 'O2', 'O3', 'O4', 'O5', 'O6', 'O7', 'O8', 'O9', 'O10']

selected_df = df[selected_columns]
```

```
corr_matrix = selected_df.corr()
corr_matrix
```

```
# Perform factor analysis
fa = FactorAnalyzer(n_factors=10, rotation=None)
fa.fit(selected_df)

# Get the eigenvalues
eigenvalues = fa.get_eigenvalues()[0]

# Plot the scree plot
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, 'bo-', linewidth=2)
plt.xlabel('Number of Factors')
plt.ylabel('Eigenvalue')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```

```
from factor_analyzer import FactorAnalyzer

# Assuming we have selected_df as the DataFrame containing the selected columns
# Perform exploratory factor analysis
efa = FactorAnalyzer(n_factors=10, rotation=None)  # Specify the number of factors want to extract
efa.fit(selected_df)

# Get the factor loadings
factor_loadings = efa.loadings_

# Display the factor loadings
print(factor_loadings)
```
